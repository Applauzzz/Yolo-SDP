```python
from ultralytics import YOLO
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import os
```

```python
model1 = YOLO("yolo11n.pt") 
results = model1.train(data="mine_detection.yaml", 
                      pretrained = True, 
                      epochs=100, 
                      imgsz=512, 
                      flipud = 0.5, 
                      translate = 0.5, 
                      degrees = 45,
                      lr0 =  0.5, 
                      lrf = 0.1)
```

### Image and Label File Preparation
Prepared as per https://docs.ultralytics.com/datasets/detect/

```python
# create train / test / validate index datasets
select = np.arange(0,12289)
select = pd.DataFrame(select)
print(select)
for i in range(len(select)):
     x = str(select.iloc[i,0])
     x = x.zfill(12)
     select.iloc[i,0] = x + '.png'
print(select)
train_select,test_select = train_test_split(select,test_size = 0.3,random_state=7)
test_select,val_select = train_test_split(test_select,test_size = 0.5,random_state=7,)
print(train_select.shape)
print(test_select.shape)
print(val_select.shape)
```

```python
#move images into train / test / validate folders.
from_path = 'D:/Downloads/Mining/'

file_list = os.listdir(from_path)

to_path = 'D:/Downloads/Mining/image version 3/images/train/'
for file in train_select[0]:
    # print(file)
    if file in file_list:
        src_path = os.path.join(from_path, file)
        dst_path = os.path.join(to_path, file)
        os.rename(src_path, dst_path)

to_path = 'D:/Downloads/Mining/image version 3/images/val/'
for file in test_select[0]:
    # print(file)
    if file in file_list:
        src_path = os.path.join(from_path, file)
        dst_path = os.path.join(to_path, file)
        os.rename(src_path, dst_path)

to_path = 'D:/Downloads/Mining/image version 3/images/test/'
for file in val_select[0]:
    # print(file)
    if file in file_list:
        src_path = os.path.join(from_path, file)
        dst_path = os.path.join(to_path, file)
        os.rename(src_path, dst_path)
```

```python
# create label index train / test / validate dataset

train_select_label = train_select.copy()
train_select_label[0] = train_select_label[0].replace(".png",".txt",regex=True)
print(train_select_label) 

test_select_label = test_select.copy()
test_select_label[0] = test_select_label[0].replace(".png",".txt",regex=True)
print(test_select_label) 

val_select_label = val_select.copy()
val_select_label[0] = val_select_label[0].replace(".png",".txt",regex=True)
print(val_select_label) 
```

```python
# Move label files to folders

from_path = 'D:/Downloads/Mining/'

file_list = os.listdir(from_path)

to_path = 'D:/Downloads/Mining/image version 3/labels/train/'
for file in train_select_label[0]:
    # print(file)
    if file in file_list:
        src_path = os.path.join(from_path, file)
        dst_path = os.path.join(to_path, file)
        os.rename(src_path, dst_path)

to_path = 'D:/Downloads/Mining/image version 3/labels/val/'
for file in test_select_label[0]:
    # print(file)
    if file in file_list:
        src_path = os.path.join(from_path, file)
        dst_path = os.path.join(to_path, file)
        os.rename(src_path, dst_path)

to_path = 'D:/Downloads/Mining/image version 3/labels/test/'
for file in val_select_label[0]:
    # print(file)
    if file in file_list:
        src_path = os.path.join(from_path, file)
        dst_path = os.path.join(to_path, file)
        os.rename(src_path, dst_path)
```

```python
# save details
train_select.to_csv("train_image.csv")
test_select.to_csv("test_image.csv")
val_select.to_csv("val_image.csv")
train_select_label.to_csv("train_label.csv")
test_select_label.to_csv("test_label.csv")
val_select_label.to_csv("val_label.csv")
```
